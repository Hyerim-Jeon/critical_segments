```python
import pandas as pd
import numpy as np

file_path = './data.csv'
data = pd.read_csv(file_path, thousands=',', header=0)
data = data.sort_values(by=['user'])
data = data.reset_index(drop=True)

# drop Question 9_2_1 (technical competency)
data.drop(data.columns[22:23], axis=1, inplace=True)

# cleansing zero (=idk) to null
data.replace({'Question_1_1_1': 0, 
            'Question_2_1_1' : 0, 
            'Question_2_2_1' : 0, 
            'Question_3_1_1' : 0,
            'Question_3_2_1' : 0, 
            'Question_4_1_1' : 0,
            'Question_4_2_1' : 0,
            'Question_4_3_1' : 0, 
            'Question_4_4_1_5_2_1' : 0,
            'Question_5_5_1' : 0, 
            'Question_6_1_1' : 0, 
            'Question_6_4_1' : 0, 
            'Question_6_5_1_5_4_1' : 0,
            'Question_6_6_1_5_1_1' : 0,
            'Question_7_1_1' : 0,
            'Question_7_2_1' : 0,
            'Question_8_1_1' : 0,
            'Question_9_1_1' : 0 
           }, '')
```

```python
# calculate 65 user's mean and std values
stv = np.empty((65,2))

user = data.loc[0,'user']
val_list = data.loc[0,data.columns[3:22]]
val = pd.Series(val_list)
uv=0

for i in range(len(data)):
    if i < len(data)-1:
        if user == data.loc[i,'user']:
            # collect value distribution
            val_list = data.loc[i,data.columns[3:22]]
            val_add = pd.Series(val_list)
            val = pd.concat([val,val_add])
        else:
            # save mean, std for user of i-1
            stv[uv][0] = val.mean(skipna=True)
            stv[uv][1] = val.std(skipna=True)

            # next user
            uv += 1
            user = data.loc[i,'user']
            val_list = data.loc[i,data.columns[3:22]]
            val = pd.Series(val_list)
    else:
        # last row
        val.astype(float)
        stv[uv][0] = val.mean(skipna=True)
        stv[uv][1] = val.std(skipna=True)
```

```python
def zscore_filter_annot(data, zthres):

    result = data.copy()
    result_2 = data.copy()
    user = result.loc[0,'user']
    uv=0

    for i in range(len(result)):
        if user == result.loc[i,'user']:
            m = stv[uv][0]
            s = stv[uv][1]
            result.loc[i,result.columns[3:22]] = (result.loc[i,result.columns[3:22]] - m ) / s
            result_2.loc[i,result.columns[3:22]] = result_2.loc[i,result_2.columns[3:22]].mask(abs(result.loc[i,result.columns[3:22]]) <= zthres, np.nan)

        else:
            # next user
            uv += 1
            user = data.loc[i,'user']

            m = stv[uv][0]
            s = stv[uv][1]
            result.loc[i,result.columns[3:22]] = (result.loc[i,result.columns[3:22]] - m ) / s
            result_2.loc[i,result.columns[3:22]] = result_2.loc[i,result_2.columns[3:22]].mask(abs(result.loc[i,result.columns[3:22]]) <= zthres, np.nan)
    
    file_name = f"zscore_result.csv"
    result.to_csv(file_name, index=False)
    
    file_name = f"zthres={zthres}_result.csv"
    result_2.to_csv(file_name, index=False)
```

```python
# second parameter = z score threshold. outlier z-score value will be bigger than this threshold

zscore_filter_annot(data, 2.5)
zscore_filter_annot(data, 2)
zscore_filter_annot(data, 1.5)
zscore_filter_annot(data, 1)
```
